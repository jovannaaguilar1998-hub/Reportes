# Introducción

Un algoritmo es una secuencia de pasos computacionales que transforma una entrada en una salida (Cormen et al., 2009). El análisis de algoritmos permite comparar la eficiencia de distintos métodos, describiendo cómo aumenta el costo de recursos (número de operaciones y tiempo de ejecución) conforme se incrementa el tamaño de entrada $n$.

En este reporte se abordan dos actividades: (i) la codificación matemática de una máquina de Turing mediante la definición formal de sus componentes y función de transición, y (ii) la comparación mediante simulación de distintos órdenes de crecimiento para estimar el tiempo de ejecución y número de operaciones bajo diferentes tamaños de entrada. El apartado experimental de este reporte fue realizado con el programa Julia.

![Figura 1. Máquina de Turing](Maquina_de_turing.png)


   

Considerando el diagrama anterior se puede definir la máquina de Turing como: 

$$M=(\{q_0,q_1,q_3,q_4,q_5,q_7,q_8,q_H\},\{1\},\{1,X,\#,\epsilon\},q_0,\epsilon,\{q_H\},\delta)$$

Por lo tanto, el **conjunto de estados queda** está definido como:

$$Q=(\{q_0,q_1,q_3,q_4,q_5,q_7,q_8,q_H\}$$

donde: 

- $q_0$: Ir al final.
- $q_1$: Volver al inicio.
- $q_2$: Escanear.
- $q_3$: Ir a la salida.
- $q_4$: Escribir 1°.
- $q_5$: Escribir 2°.
- $q_7$: Retornar.
- $q_8$: Sumar+1.
- $q_H$: Halt(Terminado).

El **alfabeto de entrada** está dado por: $\Sigma = \{1\}$

El **alfabeto de la cinta** es: $\Gamma=\{1, X, \#, \epsilon\}$

El **estado incial**: $s=q_0$

El **símbolo blanco** que representa el vacío en la máquina es: $\epsilon$ (en la *fig. 1* se presenta como *_*).

El **estado de aceptación** es: $F=\{q_H\}$

Finalmente $\delta$, la función de transición queda definida como:

$$\delta(q_0,1)=(q_0,1,R)$$
$$\delta(q_0,\epsilon)=(q_1,\#,L)$$
$$\delta(q_1,1)=(q_1,1,L)$$
$$\delta(q_1,\epsilon)=(q_2,\epsilon,R)$$
$$\delta(q_2,X)=(q_2,X,R)$$
$$\delta(q_2,1)=(q_3,X,R)$$
$$\delta(q_2,\#)=(q_8,\#,R)$$
$$\delta(q_3,1)=(q_3,1,R)$$
$$\delta(q_3,\#)=(q_3,\#,R)$$
$$\delta(q_3,X)=(q_3,X,R)$$
$$\delta(q_3,\epsilon)=(q_4,1,R)$$
$$\delta(q_4,\epsilon)=(q_5,1,R)$$

# Actividad 2. Comparación de los órdenes de crecimiento

2. Comparar mediante simulación en un notebook de Jupyter o Quarto los siguientes órdenes de crecimiento:
    - $O(1)$ vs $O(\log n)$
    - $O(n)$ vs $O(n \log n)$
    - $O(n^2)$ vs $O(n^3)$
    - $O(a^n)$ vs $O(n!)$
    - $O(n^n)$ vs $O(n!)$

## $O(1)$ vs $O(\log n)$

```julia
using Plots, LaTeXStrings

#Tamaño de entrada
n = 100

plot(1:n,[1 for x in 1:n], label=L"O(1)", title="Gráfica 1. Órdenes de crecimiento", xlabel="n", ylabel="f(n)", color= "navyblue")
plot!(1:n,[log2(x) for x in 1:n], label=L"O(\log n)", color= "chocolate")
```

**Discusión:**

La línea azul que corresponde a $O(1)$ que representa un algoritmo cuyo número de operaciones no se ve afectado al incrementar el tamaño de la entrada $n$, por lo que costo se mantiene constante (igual a 1). Este comportamiento se confirma en la tabla 1, donde el tiempo asociado a $O(1)$ no cambia al variar los tamaños de entrada propuestos en el ejercicio. 

Por su parte, la línea naranja $O (log_2 (n))$ muestra que el número de operaciones sí incrementa si el tamaño de entrada aumenta $n$.  Sin embargo, es importante señalar que este crecimiento es lento, lo cual se observa en la gráfica porque la curva se va aplanando conforme $n$ aumenta.

Al comparar ambos, podemos concluir que se trata de órdenes de crecimiento altamente eficientes dado que su costo no aumenta significativamente por incrementos en $n$. No obstante, $O(1)$ es el más eficiente, ya que se me mantiene constante al no depender de $n$. 

## $O(n)$ vs $O(n \log n)$

```julia
using Plots, LaTeXStrings

n = 100
plot(1:n, [x for x in 1:n], label=L"O(n)", title="Gráfica 2. Órdenes de crecimiento", xlabel="n", ylabel="f(n)", color=:green4)
plot!(1:n, [x * log2(x) for x in 1:n], label=L"O(n\log n)", color=:indigo)
```

**Discusión:** 

El orden de crecimiento $O(n)$ respresentado por la línea morada muestra que si se duplica el tamaño de la entrada, se duplican en número de operaciones. En consecuencia, un incremento en $n$, produce un aumento de tiempo de ejecución de forma proporcional. 

Por su parte, el orden $O(n\log n)$, representado por la curva verde, también crece de manera casi lineal, pero con un incremento adicional debido al factor $\log n$. Esto implica que, si se duplica el tamaño de entrada $n$, el número de operaciones incrementa en un poco más del doble.

Al comparar $O(n)$ con $O(n\log n)$ se observa que ambos crecen de acuerdo con el tamaño de entrada $n$, pero $O(n\log n)$ tiene un costo más elevado de operaciones (y, por lo tanto, de tiempo de ejecución) debido al factor adicional $\log n$, aunque crece de manera lenta. Por lo que $O(n)$ es un orden de crecimiento más eficiente, aunque  $O(n\log n)$ sigue siendo eficiente respecto a otros órdenes de crecimiento polinomiales. 

## $O(n^2)$ vs $O(n^3)$

```julia
using Plots, LaTeXStrings

n = 100

plot(1:n, [x^2 for x in 1:n], label=L"O(n^2)", title="Gráfica 3. Órdenes de crecimiento", xlabel="n", ylabel="f(n)", color="darkslategray3")
plot!(1:n, [x^3 for x in 1:n], label=L"O(n^3)", color="mediumorchid2")
```

**Discusión:** 

La gráfica muestra la línea azul correspondiente al orden de crecimiento $O(n^2)$. Se observa que el número de operaciones se incrementa de manera cuadrática respecto al número de entrada $n$. Por lo que si se duplica $n$, el costo se multiplica por 4, ya que $(2n)^2= 4n^2$.

Por su parte, $O(n^3)$, representado por la curva rosa, muestra un crecimiento cúbico respecto del costo de operaciones y tiempo de ejecución conforme $n$ crece. Esto implica que si se duplica el número de entrada, el costo se multiplica por 8, debido a que $(2n)^3= 8n^3$. 

En contraste, al evaluar $O(n^2)$ y $O(n^3)$, se encuentra que en ambos incrementan su costo de operaciones y tiempo de manera considerable cuando el tamaño de entrada $n$ crece.  Sin embargo, $O(n^3)$ crece más rápido su costo, por lo que para tamaños de entrada grandes, resulta menos  eficiente en comparación con $O(n^2)$. 

## $O(a^n)$ vs $O(n!)$

```julia
using Plots, LaTeXStrings

a = 2
n = 20

plot(1:n,[a^x for x in 1:n], label=L"O(a^n)",title="Gráfica 4. Órdenes de crecimiento", xlabel=L"n", ylabel=L"f(n)", color ="maroon")
plot!(1:n,[factorial(x) for x in 1:n], label=L"O(n!)", color ="lightseagreen")
```

**Discusión:** 

La línea roja muestra el orden de crecimiento exponencial $O(a^n)$, se observa que conforme crece el tamaño de entrada $n$, el número de operaciones del algoritmo aumenta en función de una base $a$, por lo tanto, el costo de aumenta de forma rápida. 

Por su parte, el $O(n!)$ representado por línea azul, refleja que un incremento en el tamaño de entrada $n$ produce un aumento muy rápido en el número de operaciones, y por lo tanto, en el tiempo de ejecución. Esto se aprecia en la gráfica incluso aunque haya un $n$ relativamente pequeño, dado que $n!$ se define como el producto $1*2*...*n$.

Al comparar ambos órdenes $O(a^n)$ y $O(n!)$, se concluye que ambos crecen rápidamente conforme aumenta el orden de entrada $n$. Sin embargo, $O(n!)$, incrementa de manera mucho más rápida el número de operaciones del algoritmo, ya que al aumentar $n$, los factores del producto se vuelven cada vez más grandes. Mientras que $O(a^n)$, siempre se multiplica por la misma base $a$, por lo que la diferencia entre ambos crecimientos se vuelve cada vez mayor. Por esta razón, un orden de crecimiento $O(n!)$ resulta muy ineficiente para tamaños de entrada incluso relativamente pequeños, en comparación con $O(a^n)$ .

## $O(n^n)$ vs $O(n!)$

```julia
using Plots, LaTeXStrings

n = 10

plot(1:n, [x^x for x in 1:n], label=L"O(n^n)", title="Gráfica 5. Órdenes de crecimiento", xlabel=L"n", ylabel=L"f(n)", color="gold3")
plot!(1:n, [factorial(x) for x in 1:n], label=L"O(n!)", color="red")
```

**Discusión:** 

La línea roja representa el orden de crecimiento $O(n!)$. Como se mencionó en la comparación anterior, se observa un crecimento del número de operaciones muy acelerado con conforme incrementa el tamaño de la entrada $n$.

Por su parte, la línea dorada muestra el orden de crecimiento $O(n^n)$, donde el número de operaciones incrementa de forma extremadamente rápida conforme aumenta el tamaño de entrada $n$, debido a que $n^n$ corresponde a multiplicar $n$ por sí mismo $n$ veces.

Al comparar los órdenes de crecimiento, observamos que $O(n^n)$ crece mucho más rápido que $O(n!)$ al aumentar el tamaño de entrada $n$. Esto se debe a que $n^n$ multiplica siempre por $n$ cada factor, mientras de que $n!$ incluye factores desde 1 hasta $n$. Como resultado, la diferencia entre ambos órdenes se hace cada vez mayor conforme $n$ aumenta. Podemos concluir ambos órdenes son altamente costosos y resultan ineficientes para tamaños de entrada grandes o moderados, donde el orden $O(n^n)$ es el más ineficiente de los dos.

```julia
using DataFrames

costo = 1e-9
sizes = [100, 1000, 10000, 100000]

orden_crecimiento = [
    ("O(1)",       n -> 1.0),
    ("O(log n)",  n -> log2(n)),
    ("O(n)",       n -> n),
    ("O(n log n)",n -> n * log2(n)),
    ("O(n^2)",     n -> n^2),
    ("O(n^3)",     n -> n^3)
]

rows = DataFrame(Orden=String[], n=Int[], tiempo=Float64[])

for n in sizes, (name, f) in orden_crecimiento
    t = f(n) * costo
    push!(rows, (string(name), n, t))
end
tabla = unstack(rows, :Orden, :n, :tiempo)
```

```julia
using DataFrames
costo = 1e-9
sizes_explosivos = [5, 10, 15, 20]
a = 2

orden_crecimiento_exp = [
    ("O(a^n)", n -> a^n),
    ("O(n!)",  n -> factorial(big(n))),
    ("O(n^n)", n -> big(n)^n)
]

rows = DataFrame(Orden=String[], n=Int[], tiempo=Float64[])

for n in sizes_explosivos
    for (name, f) in orden_crecimiento_exp
        t_exp = float(f(n)) * costo
        push!(rows, (string(name), n, t_exp))
    end
end

tabla_2 = unstack(rows, :Orden, :n, :tiempo)
```

# Conclusión

La actividad de la Máquina de Turing permitió reforzar la idea de que un algoritmo puede describirse formalmente mediante estados y una función de transición $\delta$, es decir, como un conjunto de reglas que transforman la entrada paso a paso hasta llegar a un resultado.

En general, las simulaciones analizadas muestran que mientras más alto es el orden de crecimiento, más incrementa el tiempo cuando aumenta el tamaño de entrada $n$. Los órdenes $O(1)$ y $O(\log n)$ prácticamente no se ven afectados por incrementos grandes en $n$, y $O(n)$ y $O(n\log n)$ siguen teniendo tiempos razonables para los tamaños propuestos. Sin embargo, a partir de $O(n^2)$ y $O(n^3)$ el costo se vuelve rápidamente elevado. Finalmente, los órdenes explosivos $O(a^n)$, $O(n!)$ y $O(n^n)$ crecen tan rápido que incluso con tamaños de entrada relativamente pequeños $n$, presentan tiempos de ejecución significativos, por lo que resultan poco prácticos. 
En conclusión, estos experimentos muestran que, en la práctica, es preferible usar algoritmos con órdenes de crecimiento bajos y evitar aquellos con crecimiento polinomial alto o explosivo, ya que su costo aumenta rápidamente cuando $n$ crece.
